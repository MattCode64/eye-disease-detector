```python
# creation d'un dossier pour chaque maladie
```

```python
for title in data.columns[1:]:
    if not os.path.exists(f'{title}'):
        os.mkdir(f'../../data/eyes-dataset/Sort/{title}')
        print(f"Created folder '{title}'.")
    else:
        print(f"Folder '{title}' already exists.")
```
