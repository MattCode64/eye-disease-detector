```python
import pandas as pd
from PIL import Image
import os
```

```python
# every path
train_path = r"C:\Data\Projet CODE\Code Python\DataCamp Code\data\preparing_model\RFMiD_Training_Labels.csv"
test_path = r"C:\Data\Projet CODE\Code Python\DataCamp Code\data\preparing_model\RFMiD_Testing_Labels.csv"
val_path = r"C:\Data\Projet CODE\Code Python\DataCamp Code\data\preparing_model\RFMiD_Validation_Labels.csv"

folder_images_train = r"C:\Data\Projet CODE\Code Python\DataCamp Code\data\preparing_model\images_train"
folder_images_test = r"C:\Data\Projet CODE\Code Python\DataCamp Code\data\preparing_model\images_test"
folder_images_val = r"C:\Data\Projet CODE\Code Python\DataCamp Code\data\preparing_model\images_val"

# Show every path
print(train_path)
print(test_path)
print(val_path)
```

```python
# Read csv files
train = pd.read_csv(train_path)
test = pd.read_csv(test_path)
val = pd.read_csv(val_path)
```

```python
# Shapes of dataframes
print(train.shape)
print(test.shape)
print(val.shape)
```

```python
# Values counts of 'Disease_Risk' column
print(train['Disease_Risk'].value_counts())
print(test['Disease_Risk'].value_counts())
print(val['Disease_Risk'].value_counts())
```

```python
# Read each image of the train folder and print the resolution of eahc image

# Ce code affiche toutes les résolutions de chaque image en fonction du dossier
# Parcourir tous les fichiers dans le dossier
for filename in os.listdir(folder_images_test):
    if filename.lower().endswith('.png'):
        # Construire le chemin complet de l'image
        file_path = os.path.join(folder_images_test, filename)

        # Ouvrir l'image et obtenir sa résolution
        with Image.open(file_path) as img:
            width, height = img.size

        # Afficher le nom et la résolution de l'image
        print(f"Image {filename} : {width}x{height}")
```

```python
ids = []
resolutions = []

# Parcourir tous les fichiers dans le dossier
for filename in os.listdir(folder_images_train):
    if filename.lower().endswith('.png'):
        # Enregistrer le nom du fichier sans l'extension
        id_ = os.path.splitext(filename)[0]
        ids.append(id_)

        # Construire le chemin complet de l'image et obtenir sa résolution
        file_path = os.path.join(folder_images_train, filename)
        with Image.open(file_path) as img:
            resolutions.append(f"{img.size[0]}x{img.size[1]}")

# Créer un DataFrame avec les informations
df_see = pd.DataFrame({'ID': ids, 'Resolution': resolutions})
# convert ID column to int
df_see['ID'] = df_see['ID'].astype(int)
# reset index
df_see = df_see.sort_values(by='ID').reset_index(drop=True)
```

```python
# Values counts of 'Resolution' column
print(df_see['Resolution'].value_counts())

# put in a list all the ID of the images with a resolution of 4288x2848
list_id = df_see[df_see['Resolution'] == '4288x2848']['ID'].tolist()

print(list_id)
```

```python
# put list_id in a dataframe and save it in a csv file
df_id = pd.DataFrame(list_id, columns=['ID'])
df_id.to_csv(r"C:\Data\Projet CODE\Code Python\DataCamp Code\data\preparing_model\CSV files\ID_to_DEL.csv", index=False)
```

```python
# remove of the dataframe train all the images with a resolution of 4288x2848
df_train = train[~train['ID'].isin(list_id)]
```

```python
# Check if rows have been deleted
print(df_train.shape)
```

```python
# Values counts of 'Disease_Risk' column
print(df_train['Disease_Risk'].value_counts())
```

```python
# Remove randomly 981 rows of the dataframe df_train which have a 'Disease_Risk' value of 1
df_train = df_train.drop(df_train[df_train['Disease_Risk'] == 1].sample(n=981, random_state=1).index)
```

```python
# Check if rows have been deleted
print(df_train.shape)

# Values counts of 'Disease_Risk' column
print(df_train['Disease_Risk'].value_counts())

# List of ID in the dataframe df_train
list_id_train = df_train['ID'].tolist()
```
