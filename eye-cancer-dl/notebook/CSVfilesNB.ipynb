# DATA EXPLORATION

```python
# Subplots of the features of the validation, train and test set sort by the number of '1' values
def plot_dataframes(val, test, train):
    fig, axes = plt.subplots(3, 1, figsize=(20, 15))

    dataframes = [val, test, train]
    titles = ["Validation set", "Test set", "Train set"]

    for i, df in enumerate(dataframes):
        sorted_sums = df.sum().sort_values(ascending=False)[2:]
        col = sorted_sums.index
        values = sorted_sums.values

        sns.barplot(x=col, y=values, ax=axes[i], palette='magma')
        axes[i].set_title(titles[i])
        axes[i].set_xlabel("Features")
        axes[i].set_xticks(range(len(col)))
        axes[i].set_xticklabels(col, rotation=90)

    plt.tight_layout()
    plt.show()
```

```python
plot_dataframes(val_df, test_df, train_df)
```

```python
# Check that all features have '0' and strictly '1' values
def check_values(val, test, train):
    dataframes = [val, test, train]
    titles = ["Validation set", "Test set", "Train set"]

    for i, df in enumerate(dataframes):
        for col in df.columns:
            if df[col].value_counts().shape[0] != 2:
                print(titles[i], ":", df[col].value_counts(), "\n")
```

```python
check_values(val_df.drop(['ID', 'Disease_Risk'], axis=1), test_df.drop(['ID', 'Disease_Risk'], axis=1), train_df.drop(['ID', 'Disease_Risk'], axis=1))
```

```python
# Subplot of pie chart to show the disease that come up the most in the train, validation and test set
def plot_pie_chart(val, test, train):
    fig, axes = plt.subplots(1, 3, figsize=(20, 15))

    dataframes = [val, test, train]
    titles = ["Validation set", "Test set", "Train set"]

    for i, df in enumerate(dataframes):
        df = df.drop(['ID', 'Disease_Risk'], axis=1)
        df = df.sum().sort_values(ascending=False)
        col = df.index
        values = df.values

        axes[i].pie(values, labels=col, autopct='%1.1f%%', shadow=True, startangle=90)
        axes[i].set_title(titles[i])

    plt.tight_layout()
    plt.show()
```

```python
plot_pie_chart(val_df, test_df, train_df)
```

### We see that the disease that come up the most in the train, validation and test set are the same and are perfectly balanced. But the disease themselves are not balanced. We can use the train set to train our model and the validation set to test it.

```python
# Drop the columns 'ODPM' and 'HR' of the train, test and validation set
val_df = val_df.drop(['ODPM', 'HR'], axis=1)
test_df = test_df.drop(['ODPM', 'HR'], axis=1)
train_df = train_df.drop(['ODPM', 'HR'], axis=1)
```
