# This notebook was created to sort the images of the eyes dataset into the right folders for training, testing and validation. It also deletes the images that are not in the right resolution.

### It was used to a first version of the project, but it is not used anymore. The code is kept for reference.

```python
from PIL import Image
import os
import shutil
```

```python
def count_images_by_resolution(folder_path):
    if not os.path.exists(folder_path):
        print(f"Le dossier {folder_path} n'existe pas.")
        return

    files = os.listdir(folder_path)
    resolution_count = {}

    for file in files:
        if not file.lower().endswith('.png'):
            continue
            
        file_path = os.path.join(folder_path, file)
        if os.path.isfile(file_path):
            try:
                with Image.open(file_path) as img:
                    resolution = img.size  
                    resolution_count[resolution] = resolution_count.get(resolution, 0) + 1
            except Exception as e:
                print(f"Impossible d'ouvrir l'image {file}. Erreur: {e}")

    print(f"Résolutions pour le dossier {folder_path}:")
    for resolution, count in resolution_count.items():
        print(f"Il y a {count} images de résolution {resolution}.")
    print("\n")  # Imprimez une ligne vide pour séparer les résultats des différents dossiers

# Liste des dossiers sources
source_folders = [
    "C:\Data\Projet CODE\Code Python\DataCamp Code\data\eyes-dataset\Training_Set\Training_Set\Training",
    "C:\Data\Projet CODE\Code Python\DataCamp Code\data\eyes-dataset\Test_Set\Test_Set\Test",
    "C:\Data\Projet CODE\Code Python\DataCamp Code\data\eyes-dataset\Evaluation_Set\Evaluation_Set\Validation"
]

# Appel de la fonction pour chaque dossier
for source_folder in source_folders:
    count_images_by_resolution(source_folder)
```

```python
# Spécifiez le dossier source et le dossier de destination
# source_folder = r"C:\Data\Projet CODE\Code Python\DataCamp Code\data\eyes-dataset\Training_Set\Training_Set\Training"
# dest_folder = r"C:\Data\Projet CODE\Code Python\DataCamp Code\data\eyes-dataset\Training_Set\Training_Set\testtstts"

# source_folder = r"C:\Data\Projet CODE\Code Python\DataCamp Code\data\eyes-dataset\Test_Set\Test_Set\Test"
# dest_folder = r"C:\Data\Projet CODE\Code Python\DataCamp Code\data\eyes-dataset\Test_Set\Test_Set\frefrefre"

source_folder = r"C:\Data\Projet CODE\Code Python\DataCamp Code\data\eyes-dataset\Evaluation_Set\Evaluation_Set\Validation"
dest_folder = r"C:\Data\Projet CODE\Code Python\DataCamp Code\data\eyes-dataset\Evaluation_Set\Evaluation_Set\frfzeaa"

target_resolutions = [(2144, 1424), (2048, 1536)]

# Liste pour stocker les numéros des noms de fichiers traités
file_numbers = []

# Parcourir chaque fichier dans le dossier source
for filename in os.listdir(source_folder):
    if filename.endswith(".png"):
        filepath = os.path.join(source_folder, filename)

        # Ouvrir l'image et obtenir ses dimensions
        with Image.open(filepath) as img:
            width, height = img.size

            # Vérifier si la résolution de l'image est l'une des résolutions cibles
            if (width, height) in target_resolutions:
                # Afficher le nom du fichier
                print(f"Traitement de l'image: {filename}")

                # Extraire le numéro du nom du fichier et l'ajouter à la liste
                file_number = filename.split(".")[0]  # séparer le nom du fichier et l'extension
                file_numbers.append(int(file_number))

                # Copier l'image dans le dossier de destination
                dest_filepath = os.path.join(dest_folder, filename)
                shutil.copy(filepath, dest_filepath)

print("Images triées et copiées pour les résolutions cibles du dossier source vers le dossier de destination.")
print(f"Dossier source: {source_folder}\nDossier de destination: {dest_folder}")
```
